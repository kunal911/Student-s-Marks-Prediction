Task -1: Predicting the percentage of a student based on the no.of study hours




Importing the libraries and reading the file.

```
import pandas as pd
import numpy as np

df = pd.read_csv('http://bit.ly/w-data')
df.head(15)
```

Ploting the score using scatter plot.

```
import matplotlib.pyplot as plt
%matplotlib inline
df.plot(x = 'Hours',y = 'Scores', style = 'p')
plt.xlabel('Hours studied')
plt.ylabel('Achieved score')
plt.title('Hours & Score')
plt.show
```

Splitting the data into independant and dependant variables.

```
X = df[['Hours']]
y = df['Scores']
```

Splitting the data into train and test sets.

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Initialising Linear Regression model.

```
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
Lr.fit(X_train, y_train)
print('Training the model..')
```

```
Lr.intercept_
```

```
Lr.coef_
```

Visualising coefficients and intercepts.

```
line = Lr.intercept_+X*Lr.coef_
plt.scatter(X, y, color = 'r')
plt.plot(X, line)
plt.show
```

```
print(X_test)
```

Predicting the values from test data.

```
y_prediction = Lr.predict(X_test)
```

Dataframe with Actual values and Predicted values.

```
df = pd.DataFrame({'Actual values': y_test, 'Predicted values': y_prediction})
df
```

Predicting the score for a student who studies 9.25 hours/day.

```
hours = [[9.25]]
my_pred = Lr.predict(hours)
print("Given hours {}".format(hours))
print("Predicted output {}".format(my_pred[0]))
```

  Conclusion: For 9.25 hours of study, a student can achieve approximately 93%.
